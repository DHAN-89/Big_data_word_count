```
from __future__ import print_function
from pyspark import SparkContext
sc = SparkContext.getOrCreate()
```

```
def mychange(lines):
 lines = lines.upper()
 return(lines)
```

```
my_rec= sc.textFile("/FileStore/tables/story-3.txt")
my_rec1 = my_rec.map(mychange)
my_rec1.collect()
```

```
my_rec1.count()
```

```
def lower_clean_str(x):
  punc='!"#$%&\'()*+,./:;<=>?@[\\]^_`{|}~-'
  lowercased_str = x.lower()
  for ch in punc:
    lowercased_str = lowercased_str.replace(ch, '')
  return lowercased_str
```

```
my_rec1 = my_rec.map(lower_clean_str)
my_rec1
```

```
my_rec1.take(40)
```

```
my_rec1_flat=my_rec1.flatMap(lambda x: x.split(" "))
my_rec1_flat
```

```
my_rec1_flat.take(15)
```

```
my_rec1_filter = my_rec1_flat.filter(lambda x:x!='')
my_rec1_filter
```

```
my_rec1_filter.take(4)
```

```
my_rec1_map=my_rec1_filter.map(lambda  word:(word,1))
my_rec1_map
```

```
my_rec1_map.take(15)
```

```
my_rec1_Reduce_by_key=my_rec1_map.reduceByKey(lambda x,y:(x+y)).sortByKey()
my_rec1_Reduce_by_key
```

```
my_rec1_Reduce_by_key.take(40)
```

```
my_rec1_Reduce_by_key=my_rec1_Reduce_by_key.map(lambda x:(x[1],x[0]))
my_rec1_Reduce_by_key
```

```
my_rec1_Reduce_by_key.take(10)
```

```
my_rec1_Reduce_by_key.sortByKey(False).take(100)
```
